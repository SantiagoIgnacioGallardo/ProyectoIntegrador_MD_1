```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../data/processed/streaming_users_clean.csv')
```

```python
fig, ax = plt.subplots(figsize=(8, 4))
ax.hist(df['age'], bins=20, color='steelblue', edgecolor='white')
ax.set_title('Distribución de edad de los usuarios')
ax.set_xlabel('Edad')
ax.set_ylabel('Cantidad de usuarios')
plt.tight_layout()
plt.show()
```

```python
conteo = df['subscription_plan'].value_counts()

fig, ax = plt.subplots(figsize=(6, 4))
conteo.plot(kind='bar', color=['steelblue', 'coral', 'mediumseagreen'], ax=ax)
ax.set_title('Distribución de planes de suscripción')
ax.set_xlabel('Plan')
ax.set_ylabel('Cantidad de usuarios')
ax.tick_params(axis='x', rotation=0)
plt.tight_layout()
plt.show()
```

```python
orden = ['Básico', 'Estándar', 'Premium']

fig, ax = plt.subplots(figsize=(8, 5))
df.boxplot(column='monthly_watch_time_mins', by='subscription_plan',
           ax=ax, order=orden)
ax.set_title('Tiempo de consumo mensual por plan de suscripción')
ax.set_xlabel('Plan')
ax.set_ylabel('Minutos por mes')
plt.suptitle('')
plt.tight_layout()
plt.show()
```

```python
tabla = df.groupby(['subscription_plan', 'favorite_genre']).size().unstack(fill_value=0)
tabla = tabla.loc[['Básico', 'Estándar', 'Premium']]

tabla.plot(kind='bar', stacked=True, figsize=(10, 5), colormap='tab10')
plt.title('Géneros favoritos por plan de suscripción')
plt.xlabel('Plan')
plt.ylabel('Cantidad de usuarios')
plt.legend(title='Género', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
pivot = df.pivot_table(
    values='monthly_watch_time_mins',
    index='country',
    columns='subscription_plan',
    aggfunc='mean'
).round(0)

pivot = pivot[['Básico', 'Estándar', 'Premium']]

fig, ax = plt.subplots(figsize=(8, 6))
sns.heatmap(pivot, annot=True, fmt='.0f', cmap='YlOrRd', ax=ax)
ax.set_title('Consumo promedio mensual (min) por país y plan')
ax.set_xlabel('Plan')
ax.set_ylabel('País')
plt.tight_layout()
plt.show()
```
