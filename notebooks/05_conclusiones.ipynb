## 1. Resumen de Hallazgos (Evidencia e Interpretación)

A lo largo del proyecto, hemos sometido el dataset de la plataforma de streaming a un exhaustivo análisis.

### A. Comportamiento de Consumo
* **Evidencia:** El tiempo de visualización mensual mediano se sitúa en torno a los 758 minutos (~12.6 horas), con una larga cola de usuarios "extremos" que alcanzan los 4193 minutos. En el análisis multivariado, observamos que esta distribución se mantiene constante sin importar la edad del usuario, el país de residencia o el género de contenido favorito.
* **Interpretación:** La plataforma cuenta con un engagement "base" muy sólido y universal. No existen segmentos demográficos que consuman drásticamente más o menos que otros; el producto resulta igualmente atractivo para adolescentes que para adultos mayores.

### B. Impacto del Plan de Suscripción
* **Evidencia:** Los gráficos de densidad (violín) y los análisis de promedios cruzados demostraron que los usuarios de los planes Básico, Estándar y Premium consumen exactamente la misma cantidad de horas y generan la misma cantidad de tickets de soporte técnico. Adicionalmente, el PCA no logró encontrar clústeres separables al proyectar a los usuarios según su plan.
* **Interpretación:** Pagar un plan de mayor valor no está correlacionado con un mayor uso de la plataforma ni con mayores quejas de servicio. Los clientes Premium probablemente eligen su plan por beneficios técnicos puntuales (como resolución 4K o pantallas simultáneas) y no por tener un perfil de "súper-usuario".


## 2. Conclusión General

**La principal fortaleza de la plataforma es su extrema homogeneidad.** 
Nuestra hipótesis inicial de que el tipo de plan (Strategy A) condicionaría los gustos o el tiempo de uso fue refutada por los datos. La plataforma funciona como un ecosistema donde la base piramidal del negocio en toda Latinoamérica es el Plan Básico, pero el nivel de retención (tiempo en pantalla) es idéntico en todos los segmentos. 

Desde la perspectiva del negocio, esto es altamente positivo: **no se depende de un "nicho" específico para generar horas de visualización**, sino que la oferta de contenido resuena por igual en toda la base de clientes.


## 3. Limitaciones del Proyecto

Es fundamental destacar que el alcance de estas conclusiones está condicionado por ciertas limitaciones intrínsecas del dataset provisto:

1. **Fotografía estática:** El dataset representa una captura transversal ("foto") en un momento del tiempo. No contamos con información longitudinal (series temporales) que nos permita saber si un usuario aumentó o disminuyó su consumo a lo largo de los meses.
2. **Falta de variables técnicas operativas:** No poseemos datos clave que justifican la contratación de planes superiores, tales como: cantidad de perfiles activos por cuenta, tipo de dispositivo utilizado (Smart TV vs. Smartphone) o calidad de reproducción seleccionada.
3. **Imputación de datos faltantes:** Aunque se documentó y justificó en el notebook de limpieza, la decisión metodológica de imputar los géneros faltantes bajo la categoría "Otros" y utilizar la mediana para valores numéricos extremos limita levemente la granularidad del análisis de nicho.


## 4. Mejoras Futuras (Próximos Pasos)

Para expandir y profundizar el alcance analítico de este proyecto en una etapa posterior, proponemos:

* **Integración de telemetría de dispositivos:** Cruzar estos datos con el hardware utilizado por los usuarios para validar la hipótesis de que el plan Premium se correlaciona con la visualización en televisores de alta resolución.
* **Análisis de abandono (Churn Rate):** Incorporar fechas de alta y baja de suscripción para poder modelar qué perfiles tienen mayor probabilidad de cancelar el servicio, permitiendo pasar de un análisis descriptivo a uno predictivo.
* **Procesamiento de Lenguaje Natural (NLP):** Analizar el texto libre de los tickets de soporte (actualmente solo contamos con el recuento numérico) para clasificar los problemas más frecuentes según el país de origen.
